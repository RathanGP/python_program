
# Analysing the data from data frames

Q. what are the total no. of reported cases and deaths related to covid - 19 in italy

```python
total_cases = covid_df.new_cases.sum()
total_deaths = covid_df.new_deaths.sum()
```

```python
print(f"The number of reported cases is {total_cases} and the number of reported deaths is {total_deaths}")
```

Q. what is the overall death rate (ratio of deaths to reported cases)

```python
death_rate = covid_df.new_deaths.sum()/covid_df.new_cases.sum()
```

```python
death_rate*100
```

Q. What is the overall number of tests conducted ? A total of 935310 tests were conducted before the daily test numbers were

```python
initial_tests = 935310
total_test = initial_tests + covid_df.new_tests.sum()
```

```python
total_test
```

Q. what fraction of tests returned a positive result?

```python
positive_rate = total_cases / total_test
```

```python
print('{:.2f}% of tests in Italy led to a positive diagosis.'.format(positive_rate*100))
```

# Working with dates

```python
covid_df['date'] = pd.to_datetime(covid_df.date)
```

```python
covid_df['date']
```

```python
covid_df['year'] = pd.DatetimeIndex(covid_df.date).year
covid_df['month'] = pd.DatetimeIndex(covid_df.date).month
covid_df['day'] = pd.DatetimeIndex(covid_df.date).day
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday
```

```python
covid_df
```

```python
covid_df_may = covid_df[covid_df.month==5]# Query the row for may

covid_df_may_metrics = covid_df_may[['new_cases', 'new_deaths', 'new_tests']] #Extract the subset of colums to be aggregated

covid_may_totals = covid_df_may_metrics.sum()# get the column wise sum
```

```python
covid_may_totals
```
